```python
import numpy as np
import os
import matplotlib.pyplot as plt

path = os.getcwd() + '/P2_data'
train = np.genfromtxt(path + '/P2_train.csv', delimiter=',')

data0=[]
data1=[]

for i in range(len(train)):
    if train[i][2]==0:
        data0.append(list(train[i][0:2]))
    else:
        data1.append(list(train[i][0:2]))
        
mean_0 = np.mean(data0,axis=0)
mean_1 = np.mean(data1,axis=0)

pi_0 = float(len(data0))/len(train)
pi_1 = 1 - pi_0

cov_0 = np.cov(list(map(list, zip(*data0))),ddof=0)
cov_1 = np.cov(list(map(list, zip(*data1))),ddof=0)

inv_cov_0 = np.linalg.inv(cov_0)
inv_cov_1 = np.linalg.inv(cov_1)

y0 = np.log(pi_0)-0.5*(np.log(np.absolute(np.linalg.det(cov_0))))
y5 = np.log(pi_1)-0.5*(np.log(np.absolute(np.linalg.det(cov_1))))

test_data = np.genfromtxt(path + '/P2_test.csv', delimiter=',')


x = np.arange(-5.0, 5.0, 0.02)
y = np.arange(-7.0, 7.0, 0.02)
X, Y = np.meshgrid(x, y)
X, Y = np.mgrid[-10:10:1000j, -10:10:1000j]
```
